```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
```

```python
df = pd.read_csv("train.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
y = df['label']
```

```python
X = df.copy()
```

```python
del X['label']
```

```python
X = X/255
```

```python
X.describe()
```

```python
model = MLPClassifier(solver='adam', alpha=0.01, hidden_layer_sizes=(150), random_state=1, max_iter=500, learning_rate_init=0.0001)
```

```python
model.fit(X,y)
```

```python
pred = model.predict(X)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y,pred)
```

```python
testdf = pd.read_csv("test.csv")
```

```python
testdf.head()
```

```python
testdf = testdf/255
```

```python
pred = model.predict(testdf)
```

```python
out = pd.DataFrame({"Label":pred})
```

```python
out
```

```python
out.to_csv("out_2.csv")
```
