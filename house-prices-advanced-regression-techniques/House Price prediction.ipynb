```python
RatingsDict = {
    "Ex":5,
    "Gd":4,
    "TA":3,
    "Fa":2,
    "Po":1,
    "NA":0
}

BsmtExposureDict = {
    "Gd":4,
    "Av":3,
    "Mn":2,
    "No":1,
    "NA":0
}

BsmntFinDict = {
    "GLQ":6,
    "ALQ":5,
    "BLQ":4,
    "Rec":3,
    "LwQ":2,
    "Unf":1,
    "NA":0
}

GarageFinishDict = {
    "Fin":3,
    "RFn":2,
    "Unf":1,
    "NA":0
}

PavedDriveDict = {
    "Y":2,
    "P":1,
    "N":0
}

FenceDict = {
    "GdPrv":4,
    "MnPrv":3,
    "GdWo":2,
    "MnWw":1,
    "NA":0
}
```

```python
def applyRatings(x):
    x = RatingsDict[x]
    return x

def applyBsmtExp(x):
    x = BsmtExposureDict[x]
    return x

def applyBsmntFin(x):
    x = BsmntFinDict[x]
    return x

def applyGarage(x):
    x = GarageFinishDict[x]
    return x

def applyPaved(x):
    x = PavedDriveDict[x]
    return x

def applyFence(x):
    x = FenceDict[x]
    return x
```

```python
def preprocess(df):
    df.MSSubClass = df.MSSubClass.astype('category')
    df.ExterQual = df.ExterQual.fillna('NA')
    df.ExterQual = df.ExterQual.apply(applyRatings)
    df.ExterCond = df.ExterCond.apply(applyRatings)
    
    df.BsmtQual = df.BsmtQual.fillna('NA')
    df.BsmtQual = df.BsmtQual.apply(applyRatings)
    df.BsmtCond = df.BsmtCond.fillna('NA')
    df.BsmtCond = df.BsmtCond.apply(applyRatings)
    df.BsmtExposure = df.BsmtExposure.fillna('NA')
    df.BsmtExposure = df.BsmtExposure.apply(applyBsmtExp)
    df.BsmtFinType1 = df.BsmtFinType1.fillna('NA')
    df.BsmtFinType1 = df.BsmtFinType1.apply(applyBsmntFin)
    df.BsmtFinType2 = df.BsmtFinType2.fillna('NA')
    df.BsmtFinType2 = df.BsmtFinType2.apply(applyBsmntFin)
    
    df.HeatingQC = df.HeatingQC.apply(applyRatings)
    df.KitchenQual = df.KitchenQual.fillna('NA')
    df.KitchenQual = df.KitchenQual.apply(applyRatings)
    df.FireplaceQu = df.FireplaceQu.fillna('NA')
    df.FireplaceQu = df.FireplaceQu.apply(applyRatings)
    
    df.GarageFinish = df.GarageFinish.fillna('NA')
    df.GarageFinish = df.GarageFinish.apply(applyGarage)
    df.GarageQual = df.GarageQual.fillna('NA')
    df.GarageQual = df.GarageQual.apply(applyRatings)
    df.GarageCond = df.GarageCond.fillna('NA')
    df.GarageCond = df.GarageCond.apply(applyRatings)
    
    df.PavedDrive = df.PavedDrive.apply(applyPaved)
    df.PoolQC = df.PoolQC.fillna('NA')
    df.PoolQC = df.PoolQC.apply(applyRatings)
    df.Fence = df.Fence.fillna('NA')
    df.Fence = df.Fence.apply(applyFence)
    
    df.LotFrontage = df.LotFrontage.fillna(0)
    df.MasVnrType = df.MasVnrType.fillna("None")
    df.MasVnrArea = df.MasVnrArea.fillna(0)
    df.GarageType = df.GarageType.fillna("NA")
    
    del df['Alley']
    del df['MiscFeature']
    del df['MiscVal']
    del df['GarageYrBlt']
    
    #df = df.dropna()
    
    #del df['YrSold']
    X = df.copy()
    del X['Id']
    #del X['SalePrice']
    X = pd.get_dummies(X)
    #y = df['SalePrice']
    #return X,y
    return X
```

```python
X,y = preprocess(df)
```
